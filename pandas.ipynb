```python
info = ""
file = open("info.csv1",'w')
file.write("name,age,add\n")
n = int(input("enter n ="))
for i in range(n):
    name = input("enter n = ")
    age = int(input("enter age = "))
    add = input("enter add = ")
    info = info + f"{name},{age},{add}\n"
    
file.write(info)
file.close()
```
